```python
def parse_dict(string):
    try:
        obj = ast.literal_eval(string)
    except:
        if isinstance(string, float):
            return ''
        for x in ["'wtitle:'", "'wcompany:'", "'wcity:", "'wstate:'", "'wduration:'", "'wdescr:'", "\n", "'NONE'", ": '", "'}, {"]:
            value = string.replace(x, '')
        value = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', value)
        value = re.sub(r'[,!?;.:]+', '', value)
        value = re.sub(r'[\+\*=@#<>\(\)\[\]\^_{}\\/-]+', ' ', value)
        value = re.sub(r'"', '', value)
        value = re.sub(r'[•●✓·➢⮚~&]', '', value)
        value = re.sub(r'\d+', '', value)
        value = re.sub(r'\s+', ' ', value)
        return value.lower()
    result = []
    for job in obj.values():
        for item in job:
            for value in item.values():
                if value == 'NONE': continue
                value = value.replace("\n", "").strip().lower()
                value = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', value)
                value = re.sub(r'[,!?;.:]+', '', value)
                value = re.sub(r'[\+\*=@#<>\(\)\[\]\^_{}\\/-]+', ' ', value)
                value = re.sub(r'"', '', value)
                value = re.sub(r'[•●✓·➢⮚❖~&]', '', value)
                value = re.sub(r'\d+', '', value)
                value = re.sub(r'\s+', ' ', value)
                result.append(value)
    result = ' '.join(result)
    return re.sub(r'\s+', ' ', result)

resume1['work_experiences'].sample(1).map(parse_dict).map(remove_stopwords).iloc[0]
```

```python
for x in ['work_experiences', 'Educations', 'Certificates']:
    resume1[x] = resume1[x].map(parse_dict).map(remove_stopwords)
resume1
```

```python
def parse_list(string):
    if isinstance(string, float):
        return ''
    try:
        obj = ast.literal_eval(string)
    except:
        value = string.replace("\n", "").strip().lower()
        value = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', value)
        value = re.sub(r'[,!?;.:]+', '', value)
        value = re.sub(r'[\+\*=@#<>\(\)\[\]\^_{}\\/-]+', ' ', value)
        value = re.sub(r'"', '', value)
        value = re.sub(r'[•●✓·➢⮚❖~&★]', '', value)
        value = re.sub(r'\d+', '', value)
        value = re.sub(r'\s+', ' ', value)
        return string
    result = []
    for value in obj:
        if value == 'NONE': continue
        value = value.replace("\n", "").strip().lower()
        value = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', value)
        value = re.sub(r'[,!?;.:]+', '', value)
        value = re.sub(r'[\+\*=@#<>\(\)\[\]\^_{}\\/-]+', ' ', value)
        value = re.sub(r'"', '', value)
        value = re.sub(r'[•●✓·➢⮚❖~&★]', '', value)
        value = re.sub(r'\d+', '', value)
        value = re.sub(r'\s+', ' ', value)
        result.append(value)
    result = ' '.join(result)
    return re.sub(r'\s+', ' ', result).strip()

resume1['Skills'].sample(1).map(parse_list).iloc[0]
```

```python
resume1['Skills'] = resume1['Skills'].map(parse_list)
resume1
```

```python
def parse_info(string):
    if isinstance(string, float) or string == 'NONE': return ''
    value = string.replace("\n", "").strip().lower()
    value = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', value)
    value = re.sub(r'[,!?;.:]+', '', value)
    value = re.sub(r'[\+\*=@#<>\(\)\[\]\^_{}\\/-]+', ' ', value)
    value = re.sub(r'"', '', value)
    value = re.sub(r'[•●✓·➢⮚❖~&★]', '', value)
    value = re.sub(r'\d+', '', value)
    value = re.sub(r'\s+', ' ', value)
    value = re.sub(r'technical', '', value)
    value = re.sub(r'skills', '', value)
    return value.strip()
resume1['Additional Information'].sample(1).map(parse_info).map(remove_stopwords).iloc[0]
```

```python
resume2 = []

def process_text(string):
    string = re.sub(r'[\uf0d8\uf0f6\uf0b7\uf020\uf028\uf02a]', '', string)
    string = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', string)
    string = re.sub(r'[_·-•-/■�▪]+', ' ', string)
    string = re.sub(r'[,!?;.:]+', '', string)
    string = re.sub(r'[\+\*=@#<>\(\)\[\]\^_{}\\/-]+', ' ', string)
    string = re.sub(r'"', '', string)
    string = re.sub(r'[•●✓·➢⮚❖~&★]', '', string)
    string = re.sub(r'\d+', '', string)
    string = string.replace('\t', '').lower()
    string = re.sub(r'\s+', ' ', string)
    string = remove_stopwords(string)
    return string

batches = os.listdir('datasets/Resume_dataset2/resumes_text_15k')
for x in batches:
    files = os.listdir(f'datasets/Resume_dataset2/resumes_text_15k/{x}')
    for file in files:
        with open(f'datasets/Resume_dataset2/resumes_text_15k/{x}/{file}', 'r', encoding='utf-8') as f:
            lines = [line.strip() for line in f.readlines()]
            lines = [process_text(line).strip() for line in lines if line and len(line) > 1]
            resume2.append(lines)
```
