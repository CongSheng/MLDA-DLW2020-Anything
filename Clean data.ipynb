```python
import ast
```

```python
ast.literal_eval(resume1['work_experiences'].sample(1).iloc[0])
```

```python
def parse_dict(string):
    try:
        obj = ast.literal_eval(string)
    except:
        if isinstance(string, float):
            return str(string)
        for x in ["'wtitle:'", "'wcompany:'", "'wcity:", "'wstate:'", "'wduration:'", "'wdescr:'", "\n", "'NONE'", ": '", "'}, {", ":", "[", "]", "{", "}"]:
            string = string.replace(x, '')
        return string
    result = []
    for job in obj.values():
        for item in job:
            for value in item.values():
                if value == 'NONE': continue
                value = value.replace("\n", "").strip().lower()
                result.append(value)
    return ' '.join(result)

resume1['work_experiences'].sample(1).map(parse_dict).iloc[0]
```

```python
for x in ['work_experiences', 'Educations', 'Certificates']:
    resume1[x] = resume1[x].map(parse_dict)
resume1
```

```python
resume1['Skills'].sample(1).iloc[0]
```

```python
def parse_list(string):
    if isinstance(string, float):
        return ''
    try:
        obj = ast.literal_eval(string)
    except:
        string = string.replace('\n', '')
        return string
    result = []
    for x in obj:
        if x == 'NONE': continue
        x = x.replace('\n', '')
        result.append(x.strip().lower())
    return ' '.join(result)

resume1['Skills'].sample(1).map(parse_list).iloc[0]
```

```python
resume1['Skills'] = resume1['Skills'].map(parse_list)
resume1
```

```python
resume1['Additional Information'].sample(1).iloc[0]
```

```python
def parse_info(string):
    if isinstance(string, float):
        return ''
    string = string.replace('\n', '')
    if string == 'NONE': return ''
    return string.lower()
resume1['Additional Information'].sample(1).map(parse_info).iloc[0]
```

```python
resume1['Additional Information'] = resume1['Additional Information'].map(parse_info)
resume1
```

```python
resume1['Resume_title'].sample(1).iloc[0]
```

```python
resume1['Resume_title'] = resume1['Resume_title'].str.replace('NONE', '').str.replace('\n', '').str.lower()
resume1
```

```python
resume2 = []

def process_text(string):
    string = re.sub(r'[\uf0d8\uf0f6\uf0b7\uf020]', '', string)
    string = re.sub(r'[_·-•-/■]+', ' ', string)
    string = re.sub(r'\s+', ' ', string)
    string = string.replace('\t', '').lower()
    return string

batches = os.listdir('datasets/Resume_dataset2/resumes_text_15k')
for x in batches:
    files = os.listdir(f'datasets/Resume_dataset2/resumes_text_15k/{x}')
    for file in files:
        with open(f'datasets/Resume_dataset2/resumes_text_15k/{x}/{file}', 'r', encoding='utf-8') as f:
            lines = [line.strip() for line in f.readlines()]
            lines = [process_text(line).strip() for line in lines if len(line) > 1]
            resume2.append(lines)
```
